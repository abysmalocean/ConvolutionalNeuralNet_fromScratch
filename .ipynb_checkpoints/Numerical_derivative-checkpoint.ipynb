```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
'''
Method 1, using two adjacent points to do central difference calculation.
for dy/dx calculation, input t is x, p is y.
'''
def calderiv1(t,p):
    derivP=np.zeros_like(p)
    
    # first point
    derivP[0]=(p[1]-p[0])/(t[1]-t[0])
    # last point
    derivP[-1]=(p[-1]-p[-2])/(t[-1]-t[-2])
    # the rest part
    derivP[1:-1]=(p[2:]-p[0:-2])/(t[2:]-t[0:-2])
    return derivP

'''
Method 2, using two points with some distance, j points away, to do central difference calculation.
If j is not provided, default value is 41.
'''
def calderiv2(t,p,j=None):
    
    # Initialize
    derivP=np.zeros_like(p)
    
    if j==None:
        j=41 # default value, j = 41 points
    
    # derivative calculation for middle part of the data
    for i in range(j,len(p)-j):
        derivP[i]=((p[i]-p[i-j])/(t[i]-t[i-j])*(t[i+j]-t[i])+
                (p[i+j]-p[i])/(t[i+j]-t[i])*(t[i]-t[i-j]))/(t[i+j]-t[i-j])
        
    # derivative calulation for left and right edge, where window length is less than j. Still using central different method,
    # but with distance k, which is reducing in order to keep the window not to be out-of-range
    # Left
    k=j
    for i in range(1,j):
        k=k-1
        derivP[i]=((p[i]-p[i-k])/(t[i]-t[i-k])*(t[i+k]-t[i])+
            (p[i+k]-p[i])/(t[i+k]-t[i])*(t[i]-t[i-k]))/(t[i+k]-t[i-k])
#         print(k)
    # Right
    k=j
    for i in range(len(p)-j,len(p)-1):
        k=k-1
        derivP[i]=((p[i]-p[i-k])/(t[i]-t[i-k])*(t[i+k]-t[i])+
            (p[i+k]-p[i])/(t[i+k]-t[i])*(t[i]-t[i-k]))/(t[i+k]-t[i-k])
#         print(k)

    # first point and last point
    derivP[0]=(p[1]-p[0])/(t[1]-t[0])
    derivP[-1]=(p[-1]-p[-2])/(t[-1]-t[-2])
    
    return derivP
```

Sythetic pressure is
$$
p=\ln t
$$

Derivative is calculated as $\frac{\partial p}{\partial \ln t}$ and displayed in $log_{10} - log_{10}$ scale.

```python
## Create a synthetic pressure data, for simplicity reason, just use an log curve
t=np.linspace(0.0001,100,1000)

p=np.log(t) ## Synthetic pressure
p -=np.min(p) ## shift the origin to be 0
pn = p+ np.random.randn(len(t))*0.01  ## Add some random noise

plt.plot(t,p,label='noise-free')
plt.plot(t,pn,label='with random noise')
plt.legend(loc='lower right')
# plt.show()
plt.xlabel('Time')
plt.ylabel('Pressure')

plt.xlim(-1,100)
```

```python
logt=np.log(t)

devlogt1=calderiv1(logt,p)

devlogt2=calderiv2(logt,p)

plt.loglog(logt,devlogt1,label='method 1')
plt.loglog(logt,devlogt2,label='method 2')
plt.ylim(1e-1,1e1)
plt.legend()

plt.title('Numerical derivative with noise-free pressure')
```

```python

logt=np.log(t)

devlogtn1=calderiv1(logt,pn)

devlogtn2=calderiv2(logt,pn)

plt.loglog(logt,devlogtn1,label='method 1')
plt.loglog(logt,devlogtn2,label='method 2')
plt.ylim(1e-1,1e1)
plt.legend()

plt.title('Numerical derivative with noisy pressure')
```
